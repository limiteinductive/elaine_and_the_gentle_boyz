## Data imports and preprocessing

We get the csv from the basic_COVID function in the download_data module and convert it into a multivariate TimeSeries object.

```python
covid_df = pd.read_csv('../raw_data/basic_COVID_2021-08-23.csv', index_col=0, parse_dates=True)
basic_covid = TimeSeries.from_dataframe(covid_df)
```

We can easily plot different TimeSeries with the plot method:

```python
for dep in ['75', '92', '93']:
    basic_covid[f'hosp_{dep}'].plot()
```

There is a clear connection between the curve of death and hospitalization

```python
basic_covid['dchosp_75'].diff(n=1, periods=30).plot()
basic_covid['hosp_75'].plot()
```
